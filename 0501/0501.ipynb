```python
# 정수 1부터 n까지 더할 때 그 합이 1000보다 커지게 되는 n과 합을 구하세요 while ~ break
num = 1
total = 0
while total <1000:
    total += num
    num += 1
print(total)
```

```python
# while문을 사용해 1부터 1000까지의 정수중 3의 배수의 합을구해보세요
n = 3
total = 0
while n <= 1000:
    total += n
    n += 3
print(total)
```

```python
# Task8_Challenge_0429. 다음 요구 사항을 충족하는 프로그램을 작성하세요:

# - 사용자로부터 이름, 나이, 좋아하는 색상을 입력받습니다.
# - 입력받은 정보를 딕셔너리로 저장하고, 모든 사용자 정보를 리스트에 저장합니다.
# - 이름이 'John'인 사용자의 정보만 출력하세요.
# - 전체 사용자의 평균 나이를 계산하고 출력하세요.


li1 = []
for i in range(3):
    name = input("이름 입력 : ")
    age = int(input("나이 입력 : "))
    favorite_color = input("좋아하는 색상 입력 : ")
    dic = {'name':name,'age':age,'favorite_color':favorite_color}
    li1.append(dic)
print(li1)

for i in li1:
    if i['name'] == 'John':
        print(i)

result = 0
for i in li1:
    result += i['age']
print(result/len(li1))

# 간편하게 푸는 법
result = sum(i['age'] for i in li1)
result_average = result / len(li1)
print(f"{result_average:.2f}")
```
