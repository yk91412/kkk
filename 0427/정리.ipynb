#### 리스트 메소드
**추가**
- .append(값) : 값을 리스트의 맨 뒤에 추가해준다
- .insert(idx, 값) : 값을 해당 idx에 추가해준다

**삭제**
- .remove(값) : 값을 삭제한다
    값이 없을 때는 오류 발생
- .pop(idx) : 해당 idx에 위치한 요소를 삭제한다

    없는 idx를 삭제할 때 오류 발생

    값을 보여주면서 삭제한다

**정렬**
- .sort() : 요소를 오름차순으로 정렬해준다(리스트 자체를 내부적으로 정렬)
- .sort(reverse = True) : 요소를 내림차순으로 정렬
- sorted() : 요소를 오름차순으로 정렬해주는 함수

**** sort와 sorted 차이점
1. 메서드 vs 함수:
- sort(): 리스트 객체의 내장 메서드입니다. 즉, 리스트에서만 사용할 수 있습니다.
- sorted(): 내장 함수로, 어떤 반복 가능한(iterable) 객체에도 사용될 수 있습니다. 예를 들면 리스트, 튜플, 딕셔너리, 문자열 등에 사용할 수 있습니다.

2. 반환 값:
- sort(): 리스트를 원 위치에서(in-place) 정렬하고 None을 반환합니다. 따라서 원래의 리스트 자체가 변경됩니다. -> 반환하는 값이 없다
- sorted(): 정렬된 새로운 리스트를 반환합니다. 원래의 객체는 변경되지 않습니다.

3. 유용성:
- sort(): 리스트에서만 작동하기 때문에 리스트만 정렬할 수 있습니다.
- sorted(): 다양한 객체를 정렬할 수 있으며 결과는 항상 리스트로 반환됩니다.

###### 지역변수 와 전역변수

- 지역변수

	함수 내부에서 선언한 변수

	함수 안에서만 사용 가능

- 전역변수

	함수 밖에서 선언한 변수
    
	전체에서 사용 가능

	전역 변수의 값을 함수 안에서 조회할 수 있고 global 키워드를 통해 함수 안에서
	전역 변수의 값을 변경할 수 있지만 바람직하지 않다


```python
def exam1():
    local_variable = 10
    print(local_variable)

exam1()
# 함수밖에서는 함수 내에서 만든 변수를 사용할 수 없다
```

```python
global_variable = 100

def exam2() :
    # print(global_variable)
    global_variable = 200 #exam2() 함수 내의 지역변수
    print(global_variable)

exam2()
print(global_variable)
```

```python
# global 키워드 이용한 전역변수의 값 사용 -> 전역변수 수정
global_variable = 100 #전역변수

def exam2() :
    # print(global_variable)
    global global_variable
    global_variable = 200 #전역변수의 값 수정
    print(global_variable)

exam2()
print(global_variable)
```

```python
# 전역변수를 사용시 매개변수와 리턴값을 이용\
# -> 전역변수의 값을 이용하지만 값을 수정하진 않는다
def add_number(num1, num2):
    return num1+10, num2+10

print(add_number(10,20))

x=100
y=100

print(add_number(x,y))
print(x,y)
```

##### 모듈

*** import를 하는 방법 2가지
- from 모듈명 import *

	모듈전체를 가져오는 방법

- from 모듈명 import 함수명

	특정 함수만 골라서 가져오는 방법

- as로 별칭 사용 가능하다

```python
# random : 난수 생성과 관련된 모듈
import random
# randint(a,b) : 1이상, 5이하의 숫자중에서 랜덤한 숫자 출력
print(random.randint(1,5))

# randrange()
print(random.randrange(10))# 0~9 숫자 중 랜던함 값을 리턴
print(random.randrange(2, 10)) #2~9 숫자중 랜덤한 값을 리턴
print(random.randrange(2, 10,3))# 2,5,8

# choice(시퀀스자료형)
print(random.choice(['봄', '여름','가을','겨울']))

# sample(시퀀스 자료형, 개수) : 해당 시퀀스자료형에서 개수만큼 요소를 뽑아준다
print(random.sample(['봄', '여름','가을','겨울'],2))
```

#### 메소드

- 1.인스턴스 메소드

	첫번째 매개변수를 self로 작성하는 메소드

- 2. 클래스메소드(classmethod)

	클래스 변수를 사용하는 메소드를 의미

	첫번째 매개변수를 cls로 작성하는 메소드

	self를 사용하지 않기 때문에 인스턴스 변수에는 접근할 수 없지만 cls를 통해서
	클래스 변수에 접근 가능하다

    인스턴스나 클래스명 호출 가능

    생성된 인스턴스가 없어도 호출 가능

    @classmethod 데코레이터를 표시하고 작성

    첫번째 매개변수는 cla로 작성


- 3. 스태틱메소드(정적메소드 static method)

	self, cls를 사용하지 않고 반드시 작성해야할 매개변수가 없다

    인스턴스 또는 클래스명 호출가능

    생성된 인스턴스가 없어도 호출가능

    @staticmethod 데코레이터를 표시하고 작성

    반드시 작성해야할 매개변수가 없다


```python
print("=============클래스 메소드===========")

class Korean :
    country = "한국"

    @classmethod # trip() 메소드는 클래스메소드임을 알리는 데코레이터로시작
    def trip(cls, country): #cls는 클래스의 줄임말로 Korean을 의미
        if cls.country == country : #Korean.country의미
            print("국내 여행입니다")
        else :
            print("해외 여행입니다")

Korean.trip("한국")
Korean.trip("하와이")
```

```python
print("======스태틱 메소드======")

class Calculator :
    @staticmethod
    def add(num1, num2):
        return num1+num2
    @staticmethod
    def sub(num1, num2):
        return num1-num2

# 스태틱메소드 호출
result1 = Calculator.add(10,5)
print(result1)
print(Calculator.sub(2024,2025))
```

		class 슈퍼클래스명:

- 서브클래스의 init()

	class 슈퍼클래스 :
			def __init__(self,매개변수,매개변수):
				슈퍼클래스의 초기화변수 = 매개변수

	class 서브클래스(슈퍼클래스):
		def __init__(self,name,age,school ):
			super().__init__(name,age)
			self.school = school



```python
# 슈퍼클래스
class Person:
    def __init__(self,name):
        self.name = name

    def eat(self,food):
        print(f"{self.name}이 {food}를 먹습니다")
```

```python
# 서브 클래스
class Student(Person):
    def __init__(self, name):
        super().__init__(name)

    def study(self):
        print("공부합니다")
```

```python
# 슈퍼클래스 객체
pr1 = Person("짱구")
pr1.eat("초코비")

# 서브클래스 객체
st1 = Student("하하")
print(st1)
```
