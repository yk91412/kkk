```python
# 가장 큰 수를 출력하는 함수
def print_max(a, b, c) :
    max_val = 0
    if a > max_val :
        max_val = a
    if b > max_val :
        max_val = b
    if c > max_val :
        max_val = c
    print(max_val)

print_max(5,5,2)
```

## 입력 받은 문자열 뒤집기 **

```python
def print_reverse(str2):
    print(str2[::-1])

print_reverse("python")
```

```python
# 문자열 뒤집기 2
def print_reverse2(str3):
    str3_reverse = ''
    for i in str3:
        str3_reverse = i + str3_reverse
    print(str3_reverse)
print_reverse2("python")
```

```python
# ''.join(name)를 통해서 문자열로 변환
def print_reverse3(str4):
    li4 = list(str4)
    li4.reverse()
    str4 = ''.join(li4)
    print(str4)

print_reverse3("python")
```

## 입력 문자열을 한줄에 다섯글자씩 출력하는 함수 **

```python
def print_5xn(string):
    num1 = int(len(string)/5)
    for i in range(num1+1):
        print(string[i * 5 : i * 5 + 5])

print_5xn("안녕하시렵니까제이름은무엇일까요")
```

```python
def print_5xn(string,num):
    num1 = int(len(string)/num)
    for i in range(num1+1):
        print(string[i * num : i * num + num])

print_5xn("안녕하시렵니까제이름은무엇일까요",3)
```

## 연봉을 입력받아 월급을 계산하는 함수 1원 미만은 버림 *

```python
def calc_monthly_salary(annual_salary):
    print(int(annual_salary / 12))

calc_monthly_salary(12000000.4)
```

```python
def calc_monthly_salary(annual_salary):
    print(round(annual_salary / 12))

calc_monthly_salary(12000000.9)
# 올림 ceil 내림 floor 버림 trunc은 내장함수가 아니라 math 라이브러리가 필요하다
```

```python
import math
def calc_monthly_salary(annual_salary):
    print(math.trunc(annual_salary / 12))

calc_monthly_salary(12000000.9)
```
