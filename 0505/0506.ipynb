태어난 해, 월, 일을 입력받아 사주팔자를 보는 프로그램을 작성하시오.

사주를 보는 방법)

세 수(년,월,일)가 주어지면,  (년 - 월 + 일)에 마지막 숫자가 0이면 "대박"을 출력, 그렇지 않으면 "그럭저럭"을 출력하세요.

```python
born_year = int(input("태어난 해를 입력하세요 : "))
born_month = int(input("태어난 월을 입력하세요 : "))
born_day = int(input("태어난 일을 입력하세요 : "))
born_cal = born_year - born_month + born_day
born_cal = str(born_cal)
print(born_cal)
if born_cal[-1] == 0:
    print("대박 ㅋ")
else:
    print("그럭저럭")
```

```python
def born_cal(num1, num2, num3):
    cal = num1-num2+num3
    cal = str(cal)
    if cal[-1] == 0:
        print("대박 ㅋ")
    else:
        print("그럭저럭")

born_year = int(input("태어난 해를 입력하세요 : "))
born_month = int(input("태어난 월을 입력하세요 : "))
born_day = int(input("태어난 일을 입력하세요 : "))

born_cal(born_year,born_month,born_day)
```

```python
def born_fortune(year, month, day):
    result = year - month + day
    # 뒷숫자가 0일려면 10으로 나눌 때 나머지가 0이여야함
    if result % 10 == 0:
        print("대박")
    else:
        print("그럭저럭")

def main():
    born_year = int(input("태어난 해를 입력하세요: "))
    born_month = int(input("태어난 월을 입력하세요: "))
    born_day = int(input("태어난 일을 입력하세요: "))

    born_fortune(born_year, born_month, born_day)

if __name__ == "__main__":
    main()
```

삼각형의 3변의 길이 a, b, c가 입력으로 주어진다.

여기서 입력되는 변의 관계는 a ≤ b ≤ c 이다.

그 삼각형이 무슨 삼각형인지 출력하시오.

조건에 따라 삼각형을 출력한다.

조건)

세 변의 길이가 같은 경우 : 정삼각형

두 변의 길이가 같은 경우 : 이등변삼각형

a2 + b2 = c2일 경우(피타고라스 정리) : 직각삼각형

위의 조건에 맞지 않는 일반 삼각형일 경우 : 삼각형

삼각형이 아닐 경우 : 삼각형아님

을 출력한다.

```python
def tringle(a,b,c):
    if a== b and b==c:
        print("정삼각형")
    elif a==b or b==c or a==c:
        print("이등변삼각형")
    elif (a**2+b**2==c**2) or (c**2+b**2==a**2) or (a**2+c**2==b**2):
        print("직각삼각형")
    elif (a >= b+c) or (b >= a+c) or (c >= a+b):
        print("삼각형아님")
    else:
        print("삼각형")

def main():
    a = int(input("첫번째 변의 길이를 입력하세요 : "))
    b = int(input("두번째 변의 길이를 입력하세요 : "))
    c = int(input("세번째 변의 길이를 입력하세요 : "))

    tringle(a,b,c)

if __name__ == "__main__":
    main()
```

민호는 발명을 되게 좋아하고, 컴퓨터 프로그램도 되게 좋아한다.

어느 날 민호는 컴퓨터를 사용하던 중 휴지통이 꽉 차서 불편을 느꼈다.

그래서 휴지통이 n만큼 차면 알아서 쓰레기를 압축해주는 휴지통을 만들려고 한다.

이 때 압축하는 알고리즘은 다음과 같다.

10의 자릿수와 1의 자릿수를 서로 바꾸고, 거기에 2를 곱한다.

예) 70일 경우 14가 된다.( 70 -> 07 -> 14 )

이 알고리즘은 때로는 부작용을 일으켜 휴지통의 내용이 더 많아 질지도 모른다.

만약 이 알고리즘의 심각한 부작용으로 수치가 100이 넘는다면 100의 자릿수는 무시된다.

첫째 줄에 휴지통을 압축했을 때 양을 출력한다.

둘째 줄에 그 양이 50이하이면 GOOD 을 출력하고, 50을 넘으면 OH MY GOD 을 출력한다.

```python
# 마지막 숫자가 0이길 바란다면 10으로 나누면 되고
# 두자리 수를 바꿔야한다면
# 50 -> 05 10의 나머지 (두자리수가 되어야하므로 10을 곱한다)+ 몫
# 24 -> 42
# 15 -> 51
```

```python
def compress_trash(n):
    # 10의 자릿수와 1의 자릿수를 서로 바꾸고, 거기에 2를 곱한다.
    compressed = ((n % 10) * 10 + (n // 10)) * 2
    return min(compressed, 100)  # 만약 100을 넘으면 100으로 제한

def main():
    n = int(input("휴지통의 양을 입력하세요: "))
    compressed_amount = compress_trash(n)
    print(compressed_amount)
    if compressed_amount <= 50:
        print("GOOD")
    else:
        print("OH MY GOD")

if __name__ == "__main__":
    main()
```

```python
import random
lotto_mine = []
lott_bonus = random.randint(1,46)
for i in range(6):
    lotto_mine.append(random.randint(1,46))

lotto_result = []
lotto_bonus = random.randint(1,46)
for i in range(6):
    lotto_result.append(random.randint(1,46))
count = 0

for i in lotto_mine:
    for j in lotto_result:
        if j == i:
            count += 1

if count == 6:
    print("1등")
elif count == 5 and (lott_bonus == lotto_bonus):
    print("2등")
elif count == 5:
    print("3등")
elif count == 4:
    print("4등")
elif count == 3:
    print("5등")
else:
    print("꽝")
```

영민이는 프로그램을 이용하여 계산기를 만들려고한다.

하지만 영민이는 프로그램을 얼마 배우지 않아 어려워하고있다.

우리가 영민이를 위해 계산기 프로그램을 만들어주자.

계산 결과를 정수로 출력한다.

나눗셈일 경우 실수로 출력하되 소수 둘째자리까지 출력한다.

(0으로 나누는 경우는 입력되지 않는다.)

```python
def calcul(num1,num2):
    if num2 == 0:
        print("0으로 나눌 수 없습니다")
    else:
            add = int(num1+num2)
            print(f"{num1}+{num2}={add}")
            minus = int(num1-num2)
            print(f"{num1}-{num2}={minus}")
            multip = int(num1*num2)
            print(f"{num1}*{num2}={multip}")
            divide = num1/num2
            print(f"{num1}/{num2}={divide:.2f}")

def main():
    num1 = float(input("첫번재 수 >> "))
    num2 = float(input("두번재 수 >> "))

    calcul(num1,num2)
if __name__ == "__main__":
    main()
```
