```python
# 마피아 맞추기 게임

import random
student = ['유재석','박명수','정준하','하하','노홍철','정형돈']

print(f'''
밤이 되었습니다. 마피아를 색출해주세요.
범인의 후보는 {student}입니다.
누구를 마피아로 선택하겠습니까?
''')

mapia = random.choice(student)
choice_number = 3

# elif로 한번더 색출한다면 맞추지 못했을 때
# 마지막 멘트를 다르게 할 수 있다
while choice_number > 0:
    choice1 = input(f"마피아를 지목하세요. 남은 기회는 {choice_number}번입니다.")
    if choice1 == mapia :
        print("범인을 맞췄습니다.게임을 종료합니다.")
        break
    elif (choice1 != mapia) and choice_number >1:
        print("마피아가 아닙니다. 다시 찾아보세요")
        choice_number -= 1
    else :
        print("범인을 맞추지 못하여 게임이 종료되었습니다")
        print(f"범인은 {mapia}(이)였습니다")
        break
```

```python
import math

def is_prime(num):
    if num <= 1:  # 1보다 작거나 같은 수는 소수가 아님
        return False
    if num == 2:  # 2는 소수
        return True
    if num % 2 == 0:  # 짝수는 소수가 아님
        return False
    max_divisor = math.isqrt(num)  # 제곱근까지만 나누어보면 됨
    for i in range(3, max_divisor + 1, 2):  # 홀수만 확인
        if num % i == 0:
            return False
    return True

# 예시로 소수인지 판별할 숫자 입력
number = int(input("소수인지 판별할 숫자를 입력하세요: "))

# 소수 판별 결과 출력
if is_prime(number):
    print(f"{number}은(는) 소수입니다.")
else:
    print(f"{number}은(는) 소수가 아닙니다.")
```

```python
# 문자열 압축: 반복되는 문자의 연속된 출현 횟수를 세어 압축하는 프로그램을 작성해보세요.
# 예를 들어, "aaabbbbcccd"는 "a3b4c3d1"로 압축됩니다.
text = "aaabbbbcccddddddeeeee"
di2 = {}
for i in text:
    if i not in di2:
        di2[i] = 1
    else:
        di2[i] += 1

# 각 알파벳별 빈도가 얼마인지 dict로 구하고 리스트로 바꾸기

text_exchange = list(di2.items())

li1 = []
for i in text_exchange:
    for j in i :
        li1.append(str(j))
# 리스트로 바꾸고 문자열로 바꾸기
text_str = ''.join(li1)
print(text_str)
```

```python
def compress_string(string):
    compressed_string = ""  # 압축된 문자열을 저장할 변수
    current_char = ""       # 현재 문자
    count = 0               # 현재 문자의 출현 횟수

    for char in string:
        if char == current_char:  # 현재 문자와 이전 문자가 같으면 카운트 증가
            count += 1
        else:
            if count > 0:  # 이전 문자의 출현 횟수가 0보다 크면 압축된 문자열에 추가
                compressed_string += current_char + str(count)

            current_char = char  # 현재 문자 갱신
            count = 1             # 출현 횟수 초기화

    # 마지막 문자 처리
    if count > 0:
        compressed_string += current_char + str(count)

    return compressed_string

# 예시 문자열 압축
input_string = "aaabbbbcccd"
compressed_result = compress_string(input_string)
print(compressed_result)  # 결과 출력
```

```python
# 최대공약수(GCD) 계산하기: 두 개의 숫자를 입력받아서, 그 두 숫자의 최대공약수를 계산하는 프로그램을 작성해보세요.
# 최대공약수는 두 숫자가 공통적으로 가지는 가장 큰 약수를 의미합니다.
# 두 숫자의 나머지를 계산하고, 나머지가 0이 될 때까지 두 숫자를 계속 바꾸는 과정을 반복하여 최대공약수를 찾습니다./
```

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
        print(a,b)
    return a

# 사용자로부터 두 숫자 입력 받기
num1 = int(input("첫 번째 숫자를 입력하세요: "))
num2 = int(input("두 번째 숫자를 입력하세요: "))

# 최대공약수 계산
result = gcd(num1, num2)

# 결과 출력
print(f"{num1}과(와) {num2}의 최대공약수는 {result}입니다.")
```
